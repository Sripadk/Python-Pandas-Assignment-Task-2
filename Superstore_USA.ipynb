```python
# 7 . find out how many different differnet shipement mode that we have and what is a percentage usablity of all the 
# shipment mode with respect to dataset
df1 = df_orders.groupby('Ship Mode').agg(Count_Ship_Mode = pd.NamedAgg(column = 'Ship Mode',aggfunc = 'count' ))
df1['%usability'] = (df1['Count_Ship_Mode']/len(df_orders['Ship Mode'])) * 100
df1 
```

```python
# 11 . Gorup by region and find out which region is more profitable 
df1 = df_orders.groupby('Region')['Profit'].sum().sort_values()
df1.tail(1)
```
