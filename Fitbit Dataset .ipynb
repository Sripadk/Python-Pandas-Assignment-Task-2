```python
# 4 . Find out in this data that how many unique id's we have
len(df['Id'].unique())
```

```python
# 5 . which id is one of the active id that you have in whole dataset
df.groupby('Id')['ActivityDate'].count().reset_index(name='count').sort_values(['count'], ascending=False)
```

```python
# 6 . how many of them have not logged there activity find out in terms of number of ids
df1 = df[(df['TrackerDistance'] != 0) & (df['LoggedActivitiesDistance'] == 0)]
len(df1['Id'].unique())
# This is the number of id's who have not logged their activity distance atleast once.
```

```python
# 7 . Find out who is the laziest person id that we have in dataset
# The laiest person is on the basis on the least number of steps tracelled.
df1 = df.groupby('Id')['TotalDistance'].sum().sort_values()
df1.head(1)
```

```python
# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 
# Considering more than 2000 calories burn required per day.
# df.groupby('Id')['Calories'].mean()
df1 = df.groupby('Id').agg(Calories_avg = pd.NamedAgg(column = 'Calories', aggfunc = 'mean'))
df1[df1['Calories_avg']>2000]
```

```python
# 9. how many person are not a regular person with respect to activity try to find out those 
# first we find out the total number of activity days
len(df['ActivityDate'].unique())
# as the number is 32 then wee consider at least 50% of active days required for considering as a active person.
```

```python
# Now we find the activity days for each unique ID. and those people who have activity days more than 15 only are 50% active.
# Thus we will sort those person who are not more than 50% to mark them as a not regular persons.
df.groupby('Id')['ActivityDate'].count().sort_values()
df1 = df.groupby('Id').agg(ActivityDays_Count= pd.NamedAgg(column = 'ActivityDate',aggfunc = 'count'))
df1[df1['ActivityDays_Count']<15]
```

```python
# 10 . who is the thired most active person in this dataset find out those in pandas and in sql both .
# df1 = df.groupby('Id')['TotalDistance'].sum().sort_values()
df1 = df.groupby('Id').agg(TotalDistance_sum = pd.NamedAgg(column = 'TotalDistance',aggfunc = sum))
df2 = df1.sort_values(by = 'TotalDistance_sum', ascending=False)
df2.iloc[2]
```

```python
# 11 . who is the 5th most laziest person avilable in dataset find it out
len(df2)
df2.iloc[(len(df2)-5)]
```

```python
# 12 . what is a totla acumulati9ve calories burn for a person find out 
df1 = df.groupby('Id').agg(Calories_Burn = pd.NamedAgg(column = 'Calories', aggfunc = sum))
df1.sort_values(by = 'Calories_Burn',ascending = False)
```
